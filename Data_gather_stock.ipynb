```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# Extract data history of BTC from alpha vantage API
from alpha_vantage.cryptocurrencies import CryptoCurrencies
key = "DVT0H8E56WGAV95G"
cc = CryptoCurrencies(key=key, output_format='pandas')

data, meta_data = cc.get_digital_currency_daily(symbol='BTC', market='EUR')
```

```python
data['4b. close (USD)'].plot()
plt.tight_layout()
plt.title('Daily close value for bitcoin (BTC)')
plt.grid()
plt.show()
```

```python
data
```

```python
import datetime

d1 = datetime.datetime(2020, 12, 16)
d2 = datetime.datetime(2020, 12, 7)

data_period = data.drop(data[data.index > d1].index, axis=0)
data_period = data_period.drop(data_period[data_period.index < d2]. index, axis=0)
data_period
```

```python
plt.figure()
plt.title('Daily close value for bitcoin (BTC)')
data_period['4b. close (USD)'].plot(color="red", label="BTC")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```
