```python
from newsapi import NewsApiClient
import pandas as pd
pd.set_option('display.max_colwidth', 400)

newsapi = NewsApiClient(api_key="c564dae66d524b4c9030e941b6323fad")

all_articles = newsapi.get_everything(q="bitcoin", language="en", from_param='2020-12-09', to='2020-12-11', sort_by="popularity", page_size=100)
```

```python
df = pd.DataFrame(all_articles["articles"])
data = pd.DataFrame({"title" : df.title,
                 "description" : df.description})
```

```python
data
```

```python
data = data.drop_duplicates(subset=['description']).reset_index(drop=True)
data = data.dropna()
```

```python
# Clean
import re
import string

def clean_text(text):
    if type(text) == str:
        text = text.lower()
        text = re.sub('\[.*?\]', '', text)
        text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('[‘’“”…]', '', text)
        text = re.sub('\n', '', text)
        return text
    return text

clean = lambda x: clean_text(x)
```

```python
data_clean = data
data_clean.title = data.title.apply(clean)
data_clean.description = data.description.apply(clean)
data_clean
```
